```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
#x=입력값
x=np.arange(-5.0, 5.0, 0.1)
#y
y=sigmoid(x)
```

```python
plt.plot(x,y, 'g')
plt.plot([0,0], [4.0, 0.0], ':') #x축은 0,0 y축은 0.0~1.0까지 
plt.title('Sigmoid Function')
plt.show()
```

```python
y1= sigmoid(0.5*x)
y2= sigmoid(x)
y3= sigmoid(2*x)
plt.plot(x,y1, 'r', linestyle='--')
plt.plot(x,y2, 'g')
plt.plot(x,y3, 'b', linestyle='--')
plt.plot([0,0], [1.0, 0.0], ':')
plt.show()
```
